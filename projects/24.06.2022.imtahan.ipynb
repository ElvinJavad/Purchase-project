```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv(r'C:\Users\Acer-014\Desktop\train.csv')
```

```python
df.head()
```

```python
df['Age']=df['Age'].map(lambda row: 'teenager' if row=='0-17' else 'adult')
df=df.rename(columns={'Age': 'Age_Category'})
```

```python
df['Marital_Status']=df['Marital_Status'].map(lambda row: 'Single' if row==0 else 'Married')
```

```python
df['Product_Category_3'].fillna('not defined', inplace=True)
df['Product_Category_2'].fillna('not defined', inplace=True)
```

```python
df.head()
```

```python
df['Purchase'].value_counts().sort_values(ascending=False)[:20].plot.bar()
```

```python
purchase=df['Purchase'].value_counts()[:5]

words =purchase.index
sizes = purchase.values

plt.pie(sizes, labels=words, autopct='%1.1f%%')
plt.show()
```
